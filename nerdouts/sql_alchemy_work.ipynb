```python
from sqlalchemy import MetaData, create_engine, text, inspect
import models
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, select
from datetime import timedelta
from sqlalchemy.orm import Session
from datetime import datetime
from models import Exercise, ExerciseCategory, Base, Workout, WorkoutExercise
import random
engine = create_engine('sqlite:///:memory:')
models.Base.metadata.create_all(engine)
```

```python
exercises = [
    Exercise(
        name="Push-ups",
        category_id=ExerciseCategory.strength,
        side="None",
        default_time=30,
        repetition=10,
        prompt="Do {name}",
        long_desc="Standard push-ups working chest and triceps"
    ),
    Exercise(
        name="Hamstring Stretch",
        category_id=ExerciseCategory.stretch,
        side="Both",
        default_time=45,
        repetition=3,
        prompt="Stretch {name} on {side} side",
        long_desc="Gentle hamstring stretch while seated"
    ),
    Exercise(
        name="Foam Rolling IT Band",
        category_id=ExerciseCategory.rolling,
        side="Both",
        default_time=60,
        repetition=2,
        prompt="Roll {name} on {side} leg",
        long_desc="Use foam roller on IT band from hip to knee"
    ),
    Exercise(
        name="Ankle Mobility",
        category_id=ExerciseCategory.physical_therapy,
        side="Both",
        default_time=20,
        repetition=10,
        prompt="Do {name} exercises on {side} ankle",
        long_desc="Ankle mobility exercises to improve range of motion"
    ),
    Exercise(
        name="Squats",
        category_id=ExerciseCategory.strength,
        side="None",
        default_time=40,
        repetition=12,
        prompt="Perform {name}",
        long_desc="Full body squats with proper form"
    )
]

# Function to create a random workout
def create_random_workout(session, available_exercises, date):
    workout = Workout(created_date=date)
    session.add(workout)
    
    # Select 2-4 random exercises for this workout
    selected_exercises = random.sample(available_exercises, random.randint(2, 4))
    
    for exercise in selected_exercises:
        # Randomly adjust time and reps slightly from defaults
        time_adjustment = random.uniform(0.8, 1.2)  # ±20%
        rep_adjustment = random.randint(-2, 2)  # ±2 reps
        
        workout_exercise = WorkoutExercise(
            workout=workout,
            exercise=exercise,
            time_per_set=int(exercise.default_time * time_adjustment),
            repetition=max(1, exercise.repetition + rep_adjustment),
            created_date=date
        )
        session.add(workout_exercise)
    
    return workout

# Create sample data
with Session(engine) as session:
    # Add exercises
    for exercise in exercises:
        session.add(exercise)
    
    # Create 10 workouts over the last 30 days
    end_date = datetime.now()
    start_date = end_date - timedelta(days=30)
    
    workout_dates = sorted([
        start_date + timedelta(days=random.randint(0, 30))
        for _ in range(10)
    ])
    
    workouts = [
        create_random_workout(session, exercises, date)
        for date in workout_dates
    ]
    
    session.commit()
    
    # Print out the generated data
    print("\n=== Generated Exercises ===")
    for exercise in exercises:
        print(f"- {exercise.name} ({exercise.category_id.value})")
    
    print("\n=== Generated Workouts ===")
    for workout in workouts:
        print(f"\nWorkout {workout.id} on {workout.created_date.strftime('%Y-%m-%d')}:")
        for we in workout.workout_exercises:
            print(f"- {we.exercise.name}: {we.repetition} sets of {we.time_per_set}s")
        print(f"Total time: {workout.get_total_time}s")

# Return engine for further experimentation
print("\nDatabase populated and ready for querying!")
```
